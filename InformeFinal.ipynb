**1.2. HISTORIA, USO Y PROYECTOS FUTUROS RELACIONADOS CON BCI**
<p style="margin-left: 25px; text-align: justify;">
**1.2.1. Historia:** El desarrollo de las interfaces cerebro-computador (BCI) se origina con la electroencefalografía (EEG), introducida por Hans Berger a inicios del siglo XX, quien logró registrar la actividad eléctrica cerebral e identificar ritmos como las ondas alfa y beta, asociadas a estados de relajación y alerta [2].
</p>
<p style="margin-left: 25px; text-align: justify;">
Posteriormente, el EEG permitió plantear la posibilidad de usar señales cerebrales como canal de comunicación. En las décadas de 1950 y 1960, se demostró que los usuarios podían modular voluntariamente su actividad cerebral (neurofeedback), base para el control en BCI [2].
</p>
<p style="margin-left: 25px; text-align: justify;">
En 1973 se introdujo el término Brain-Computer Interface, y en 1988 se desarrolló el sistema P300 Speller, permitiendo comunicación mediante señales EEG. Más adelante, surgieron paradigmas como la imaginación motora y los Potenciales Evocados Visualmente en Estado Estacionario (SSVEP), que actualmente son ampliamente utilizados en sistemas BCI [2].
</p>

<p style="margin-left: 25px; text-align: justify;">
Actualmente, la BCI se define como un sistema de interacción humano-computador de lazo cerrado, cuya arquitectura técnica se fundamenta en seis componentes críticos: el usuario (SNC), el paradigma de la tarea, la codificación neural, la adquisición de señales, el procesamiento basado en computadoras y la retroalimentación en tiempo real [3]. Bajo esta perspectiva, el usuario se integra como un controlador adaptativo clave gracias a la plasticidad neuronal, mientras que los paradigmas y la codificación neural establecen los principios científicos que permiten traducir intenciones específicas en comandos digitales con precisión [3].
</p>

<p style="margin-left: 25px; text-align: justify;">
**1.2.2. Aplicaciones:**La versatilidad de las BCI ha permitido su expansión más allá del ámbito clínico, consolidándose bajo un marco de "eficacia potencial" que define cómo la tecnología interactúa con las capacidades naturales del usuario. Según la taxonomía técnica actual, estas interfaces se clasifican en cinco categorías funcionales: <b>reemplazo</b>, donde la señal artificial sustituye una función motora perdida; <b>restauración</b>, enfocada en la rehabilitación para recobrar el control biológico; <b>mejora</b>, dirigida a potenciar capacidades normales en sujetos sanos; <b>suplemento</b>, que añade nuevas modalidades de control no naturales; e <b>mejora técnica (improvement)</b> [3]. En el ámbito clínico, estas categorías se materializan principalmente en dos grandes áreas: el control de dispositivos asistivos y la neurorehabilitación [4]. <p style="margin-left: 25px; text-align: justify;">
<strong>1.2.3. Proyectos futuros y desafíos:</strong> El desarrollo de las BCI se orienta hacia la transición de sistemas experimentales a aplicaciones clínicas robustas, centradas en la inducción de neuroplasticidad adaptativa. Una tendencia emergente es la convergencia entre las BCI y la neuromodulación, donde sistemas de lazo cerrado (<i>closed-loop</i>) no solo decodifican la intención, sino que emplean técnicas como el ultrasonido o la estimulación cortical para optimizar el entrenamiento y mejorar el desempeño del usuario [3]. No obstante, persiste un debate sobre la madurez tecnológica; mientras innovaciones en hardware y robótica quirúrgica (como las propuestas por Neuralink) prometen alta resolución, su estabilidad a largo plazo y eficacia real aún requieren validación en ensayos clínicos extensos para evitar expectativas desproporcionadas en el sector médico [3].
</p>

<p style="margin-left: 25px; text-align: justify;">
En el ámbito de la ingeniería de señales, el futuro de las BCI no invasivas depende de superar las limitaciones del EEG. Los desafíos actuales incluyen el desarrollo de sensores innovadores, como los electrodos endovasculares (<i>stents</i> cerebrales), que capturan actividad de alta calidad con menor riesgo que la cirugía abierta [2]. Asimismo, se proyecta el uso de paradigmas multimodales que combinen la imaginación motora con tareas cognitivas adicionales o señales fisiológicas (expresiones faciales, fNIRS) para aumentar la robustez del sistema [2]. Finalmente, la inteligencia artificial se enfoca en resolver la escasez de datos mediante el procesamiento de señales tensoriales (<i>data completion</i>) y la generación de datos sintéticos, permitiendo entrenar clasificadores profundos más precisos incluso ante registros ruidosos o incompletos, facilitando así la futura estandarización de la tecnología en la práctica clínica [2-3].
</p>

**1.3. Actividad cerebral registrada por EEG durante movimientos e imaginación 
de movimiento de las manos y cómo se hace su análisis en sistemas BCI: Electrodos, ritmos 
cerebrales de interés**

<p style="margin-left: 25px; text-align: justify;">
La base neurofisiológica de la imaginación motora (IM) reside en la activación de redes sensorimotoras sin ejecución física. Este proceso se fundamenta en la <b>Desincronización Relacionada con Eventos (ERD)</b> y la <b>Sincronización Relacionada con Eventos (ERS)</b> [5]. En otros términos, la <b>ERD</b> es la interrupción del ritmo grupal de las neuronas (desincronización), lo que se traduce en una <b>caída de la potencia</b> en el EEG. Por el contrario, la <b>ERS</b> es el retorno al ritmo grupal coordinado (sincronización), lo que genera un <b>aumento de la potencia</b>.
</p>

<p style="margin-left: 25px; text-align: justify;">
Esta relación ocurre porque, cuando las <b>neuronas piramidales</b> de la corteza motora están en reposo, disparan todas al mismo tiempo, sumando sus voltajes y creando ondas de gran amplitud. Al iniciar la IM, estas neuronas aumentan su excitabilidad y sedespolarizan para procesar la tarea de forma individualizada; al dejar de disparar al unísono, las señales se cancelan entre sí, reduciendo la potencia total registrada. Según <b>Yu et al.</b>, este patrón es asimétrico: la ERD de mu ocurre en el <b>EEG contralateral</b> (lado opuesto a la mano), mientras que el ritmo beta suele presentar una ERS en el <b>EEG ipsilateral (mismo lado)</b> [5].
</p>

<p style="margin-left: 25px; text-align: justify;">
Para capturar estos cambios, los estudios utilizan el sistema de posicionamiento <b>10-10</b> (mencioando en el dataset que se usará) y también existe el 10-20 [5]. Por otra parte, la evidencia indica que tanto movimiento real (MR) como Movimiento imaginario(MI) 
generan ERD en las bandas Mu y Beta sobre la corteza sensorimotora contralateral [6]. 
Sin embargo, existen diferencias importantes: la ERD inducida por MR tiende a ser de 
<b>mayor magnitud y más localizada espacialmente</b>, mientras que la ERD de la IM 
es de <b>menor amplitud y más difusa</b>, lo que refleja una mayor carga cognitiva [6]. 
Esto se traduce en que los sistemas BCI alcanzan mayor precisión de clasificación con ME 
que con IM en usuarios sin entrenamiento [6]. A nivel cortical, Wang (2025) confirmó 
mediante localización de fuentes que en ambas condiciones se activan las mismas regiones 
en bandas Mu y Beta, corteza motora primaria y área somatosensorial, lo que respalda 
el uso de los mismos electrodos centrales (C3, C4) [6] y las mismas bandas frecuenciales 
para analizar tanto ME como IM en sistemas BCI [6].
</p>

**1.4. Flujos de procesamiento propuestos para estos tipos de proyectos 
y que se puedan realizar en el marco del curso.**

<p style="margin-left: 25px; text-align: justify;">
Según Raza y Yusoff [8], el procesamiento de señales EEG para sistemas BCI tiene un flujo que comprende cuatro etapas principales: preprocesamiento, extracción de características, clasificación y 
retroalimentación hacia la aplicación.
</p>

<p style="margin-left: 25px; text-align: justify;">
La primera etapa es el <b>preprocesamiento</b>, cuyo objetivo es mejorar la calidad de la señal eliminando ruido y artefactos. Zheng et al. [7] proponen una secuencia concreta aplicable al tipo de dataset (pequeño) de este proyecto: filtro pasa-altas a 1 Hz (FIR de fase cero) para eliminar la deriva de línea base, filtro notch a 50 Hz para suprimir el ruido de la red eléctrica, y filtro pasa-banda entre 4 y 40 Hz (un Chebyshev de sexto orden, muy preciso en el corte) para atenuar artefactos fisiológicos como ECG y EMG. Adicionalmente, Raza y Yusoff [8] señalan técnicas complementarias como el Análisis de Componentes Independientes (ICA) para remover ruido ocular y muscular, y la segmentación temporal (<i>epoching</i>) para sincronizar los segmentos con los marcadores de eventos T1 y T2 presentes en el dataset.
</p>

<p style="margin-left: 25px; text-align: justify;">
La segunda etapa es la <b>extracción de características</b>. En el dominio frecuencial, el método central es la <b>Densidad Espectral de Potencia (PSD)</b> calculada mediante la Transformada Rápida de Fourier (FFT) sobre las bandas Mu (8–12 Hz) y Beta (13–30 Hz) [8]. En el dominio espacial, Zheng et al. [7] demuestran que para datasets de IM de manos es posible reducir hasta un 27% el número de canales usando la <b>entropía de energía wavelet</b> como criterio de selección, conservando principalmente los electrodos de la región sensorimotora central. Dado que el dataset de 
este proyecto cuenta con 64 canales en montaje 10-10, el análisis se concentrará en <b>C3 y C4</b>, y otros cercanos a estos según el mismo autor, estos se plantearán el plan de análisis.
</p>

<p style="margin-left: 25px; text-align: justify;">
La tercera etapa es la <b>clasificación y comparación estadística</b>. El análisis estadístico descriptivo e inferencial sobre la PSD por condición (reposo, IM mano derecha, IM mano izquierda, movimiento real) es el enfoque principal. Raza y Yusoff [8] documentan que los clasificadores tradicionales como el Análisis Discriminante Lineal (LDA) y la Máquina de Vectores de Soporte (SVM) son el punto de partida estándar del campo, ofreciendo alta interpretabilidad y bajo costo 
computacional, siendo directamente implementables con scikit-learn en Python.
</p>

En el presente flujo se propone una metodología de  procesamiento y análisis que permita evidenciar la diferencia en ritmos cerebrales asociados a las cuatro condiciones de estudio en los electrodos relevantes como se plantea en el proyecto.

#### Paso 1: Carga de datos

Se descargarán los datos del data set de los 109 sujetos según los runs (tareas) de interés planteadas anteriormente. Se leerán los archivos `.set` de EEGLAB utilizando `mne.io.read_raw_eeglab()` y se analizarán las variables necesarias del formato.

- **Runs 3, 7, 11:** Tarea 1, movimiento real de mano izquierda (T1) y derecha (T2)
- **Runs 4, 8, 12:** Tarea 2, imaginación motora de mano izquierda (T1) y derecha (T2)

El análisis se concentrará en los canales de la región sensoriomotora planteados por Zheng et al. [7] en una de sus combinaciones, son estos: `Fz, Fc3, Fc1, Fc4, C5, C3, Cz, C4, Cp3, Cp1, Cpz, Cp2, Cp4, P1, P2, Poz`

---

#### Paso 2: Inspección visual pre-filtrado

Antes de aplicar cualquier filtro se realizará una inspección de la señal cruda tanto 
en el dominio temporal como en el frecuencial. En el dominio temporal se evaluará la 
presencia de deriva de línea base, oscilaciones lentas, picos transitorios y 
estabilidad general de los registros. En el dominio frecuencial se calculará la PSD mediante el método de Welch sobre la señal sin procesar, con el objetivo de 
identificar ruido de red eléctrica, componentes de muy baja frecuencia y la 
distribución de energía en las bandas funcionales de interés.

---

#### Paso 3: Filtrado

Se utilizará un filtro **IIR Butterworth** aplicado con `filtfilt` 
(`phase='zero-double'` en MNE), que procesa la señal hacia adelante y hacia atrás 
para corregir el desfase no lineal inherente a los filtros IIR. 

El Butterworth es elegido porque su banda de paso es maximamente plana, lo que 
garantiza que no se introducen distorsiones en las amplitudes de las frecuencias de 
interés. Frente al FIR, el IIR requiere un orden mucho menor para lograr la misma 
atenuación.

El orden se determina a partir de las especificaciones de atenuación requeridas para 
EEG y puede calcularse formalmente con `scipy.signal.buttord`. Al aplicar `filtfilt`, 
el orden efectivo se duplica a 8, lo que garantiza una atenuación superior a 48 dB en 
la banda de rechazo — suficiente para separar las bandas de interés. Órdenes 
superiores a 8 en IIR pueden generar inestabilidad numérica, por lo que el orden 4 
representa el balance óptimo entre selectividad y estabilidad.

**Consideración sobre la frecuencia de corte y la banda de transición:** A diferencia 
de los filtros FIR donde la frecuencia de corte se ubica en el punto de -6 dB, en los 
filtros IIR la frecuencia de corte corresponde al punto de **-3 dB**. Adicionalmente, 
al ser IIR, la transición entre banda de paso y banda de rechazo no es abrupta. Por 
ello, las frecuencias de corte se ubican con un margen de 1–2 Hz respecto a los 
límites de las bandas de interés, para garantizar que a la frecuencia deseada el 
filtro ya se encuentre en zona plana.

La secuencia de filtrado es la siguiente:

1. **Filtro pasa-altas a 0.5 Hz:** elimina la deriva de línea base y artefactos de 
muy baja frecuencia (sudoración). La frecuencia de corte se fija en 0.5 Hz en lugar 
de 1 Hz para que, considerando que el punto de -3 dB del IIR cae exactamente en la 
frecuencia de corte y la banda de transición es gradual, a 1 Hz el filtro ya opere 
en zona plana.

3. **Filtro pasa-bajas a 40 Hz:** elimina artefactos musculares y componentes de alta 
frecuencia no relevantes para el análisis motor. Se aplica con pendiente menos 
pronunciada que el pasa-altas, aplicando filtros más suaves 
en la banda superior para no afectar el contenido espectral de la banda Beta (hasta 
30 Hz). Este también elimina el rudio eléctrico por lo que no es necesario el filtro notch.

**Nota importante:** No se filtra por bandas específicas (Mu, Beta) antes del cálculo 
de la PSD, con el fin de conservar la posibilidad de normalizar la potencia de cada 
banda respecto a la potencia total del rango completo (0.5–40 Hz).

---

#### Paso 4: Inspección visual post-filtrado

Se comparará la señal antes y después de cada filtro tanto en el dominio temporal 
como en el frecuencial, verificando la eliminación del pico en 60 Hz, la reducción 
de la deriva basal y la conservación de la actividad en las bandas Mu y Beta.

---

#### Paso 5: Segmentación por eventos (Epoching)

Una vez filtradas las señales se procederá a la segmentación temporal utilizando los 
marcadores de eventos del dataset:

- **T0:** reposo
- **T1:** inicio de movimiento o imaginación de mano izquierda
- **T2:** inicio de movimiento o imaginación de mano derecha

La ventana temporal será según lo que se haya visto en la inspección inicial del 
dataset, obteniendo un arreglo de forma `(n_épocas × n_canales × n_muestras)` que 
constituye la base para el procesamiento vectorizado posterior.

---

#### Paso 6: Cálculo de la Densidad Espectral de Potencia (PSD)

Para cada época y canal de interés se calculará la PSD mediante el **método de Welch** 
(`scipy.signal.welch`), operando de forma vectorizada sobre el eje de muestras 
(`axis=-1`) para procesar simultáneamente todas las épocas y canales sin bucles 
innecesarios.

La PSD se calculará sobre el rango completo **0.5–40 Hz** y la potencia de cada banda 
se obtendrá integrando el área bajo la curva espectral mediante `numpy.trapz` sobre 
los índices de frecuencia correspondientes a cada banda:

| Banda  | Rango (Hz) |
|--------|------------|
| Delta  | 0.5 – 4    |
| Theta  | 4 – 8      |
| Mu     | 8 – 12     |
| Beta   | 13 – 30    |
| Gamma  | 30 – 40    |

Cada valor de potencia por banda se **normalizará** respecto a la potencia total del 
rango completo, calculada también por integración con `numpy.trapz` sobre el rango 
0.5–40 Hz:

$$P_{norm} = \frac{P_{banda}}{P_{total\ (0.5-40\ Hz)}}$$

Esta normalización es fundamental porque elimina diferencias absolutas de amplitud 
entre sujetos que no reflejan diferencias en activación cortical, y requiere que la 
PSD se calcule sobre la señal completa antes de extraer bandas individuales.

---

#### Paso 6.1: Cálculo de ERD/ERS

A partir de la PSD normalizada se calculará la **Desincronización/Sincronización 
Relacionada con Eventos (ERD/ERS)**, que cuantifica el cambio porcentual de potencia 
espectral en cada banda respecto al período de reposo (baseline):

$$ERD/ERS(\%) = \frac{P_{tarea} - P_{baseline}}{P_{baseline}} \times 100$$

Donde:
- $P_{tarea}$ es la potencia normalizada durante la condición de interés 
(mov_real_izq, mov_real_der, imag_izq, imag_der)
- $P_{baseline}$ es la potencia normalizada promedio de los runs 1 y 2 (reposo)

Un valor **negativo** indica **ERD** (desincronización — activación cortical), 
mientras que un valor **positivo** indica **ERS** (sincronización — inhibición 
cortical).

El ERD/ERS se calculará principalmente sobre las bandas **Mu (8–12 Hz)** y 
**Beta (13–30 Hz)** en los canales C3, C4 y Cz, por ser los marcadores 
neurofisiológicos clásicos del control motor. Las demás bandas y canales se 
reportarán como análisis complementario.

La PSD normalizada constituye la base para las pruebas estadísticas entre 
condiciones, mientras que el ERD/ERS permite la interpretación neurofisiológica 
de las diferencias observadas.

### Selección de pruebas estadísticas

Antes de seleccionar la prueba estadística adecuada, se evaluará la normalidad de los 
datos mediante las pruebas de **Shapiro-Wilk** y **Kolmogorov-Smirnov**. Se incluyen 
ambas porque con muestras grandes (109 sujetos y múltiples épocas) Shapiro-Wilk puede 
detectar como no normal cualquier desviación mínima sin relevancia práctica, por lo 
que el criterio conjunto de ambas pruebas ofrece una evaluación más robusta.

Dependiendo del comportamiento de la distribución, se procederá de la siguiente manera:

Si existe normalidad, se utilizarán pruebas paramétricas como:

- t de Student (para comparaciones entre dos grupos)
- ANOVA (para comparaciones entre más de dos grupos simultáneamente)

Si no existe normalidad, se emplearán pruebas no paramétricas como:

- Mann-Whitney U (para comparaciones entre dos grupos)
- Kruskal-Wallis (para comparaciones entre más de dos grupos, equivalente no 
paramétrico del ANOVA, aplicable especialmente al comparar las cuatro condiciones 
simultáneamente)

Dado que las señales EEG suelen presentar alta variabilidad intersujeto y 
distribuciones no normales, es frecuente que el análisis requiera pruebas no 
paramétricas.

***Filtrado***

Con el fin de mejorar la calidad de las señales EEG y facilitar la extracción de características relevantes para el análisis de imaginación motora, se implementó una etapa de filtrado digital basada en filtros IIR tipo Butterworth, aplicados mediante un esquema de fase cero (zero-phase filtering).

**Diseño del filtro**

Se seleccionó un filtro Butterworth debido a su respuesta en frecuencia maximamente plana en la banda de paso, lo que permite preservar las amplitudes de las componentes espectrales de interés sin introducir distorsiones.

A diferencia de los filtros FIR, los filtros IIR permiten alcanzar niveles de atenuación adecuados con un orden significativamente menor, lo que reduce la complejidad computacional. Sin embargo, los filtros IIR presentan desfase no lineal; por esta razón, se empleó la función filtfilt, la cual aplica el filtro en sentido directo e inverso, eliminando el desfase y duplicando el orden efectivo del sistema.

El orden del filtro fue determinado a partir de especificaciones formales de diseño mediante scipy.signal.buttord, considerando:

* Frecuencia de paso: 1 Hz
* Frecuencia de rechazo: 0.5 Hz
* Atenuación en banda de paso: 3 dB
* Atenuación en banda de rechazo: 40 dB

A partir de estas condiciones, se obtuvo un orden mínimo de N=7. No obstante, con el fin de garantizar estabilidad numérica, se implementó un filtro de orden 4 en representación SOS (Second-Order Sections), evitando problemas de inestabilidad asociados a coeficientes elevados en implementaciones directas.

Dado que el filtrado se realiza mediante filtfilt, el orden efectivo del sistema es 8 (o 14 en el caso de diseño inicial), lo que asegura una atenuación superior a 48 dB en la banda de rechazo.

El desempeño del filtro fue evaluado mediante el análisis de su respuesta en frecuencia utilizando sosfreqz, considerando tres puntos clave: banda de rechazo, frecuencia de corte y banda de paso.

Los resultados obtenidos evidencian:

**Filtro pasa-altas (0.5 Hz):**
* Atenuación de −198.60 dB a 0.1 Hz → eliminación efectiva de la deriva DC.
* Atenuación de −6.26 dB en la frecuencia de corte → consistente con doble aplicación (filtfilt).
* Atenuación cercana a 0 dB a 1 Hz → preservación de la señal útil.

**Filtro pasa-bajas (40 Hz):**
* Atenuación de −0.03 dB a 30 Hz → preservación de la banda beta.
* Atenuación de −6.02 dB en la frecuencia de corte.
* Atenuación de −107.18 dB a 60 Hz → eliminación efectiva del ruido de red eléctrica.

Estos resultados confirman que el filtro cumple y supera los criterios de diseño establecidos.

### Discusión y analisís

***Etapa Pre Filtrado***

El análisis de las señales EEG en estado crudo permitió evidenciar diferencias y limitaciones inherentes a la adquisición de datos neurofisiológicos, así como establecer una base para justificar las decisiones de procesamiento posteriores.

En ambas condiciones (movimiento real e imaginación motora), las señales crudas presentan una alta variabilidad temporal, con oscilaciones complejas y superposición de múltiples componentes de distinta naturaleza. Este comportamiento refleja la suma de actividad neuronal sincronizada junto con diversas fuentes de ruido fisiológico y no fisiológico.

Se observa la presencia de:

* Componentes de muy baja frecuencia, asociados a deriva de línea base, movimientos lentos o cambios en la impedancia de los electrodos.
* Fluctuaciones irregulares en amplitud, que dificultan la identificación directa de patrones relacionados con la actividad motora.
* Ausencia de una estructura claramente diferenciable entre movimiento real e imaginación motora a simple inspección visual.

El análisis de la densidad espectral de potencia (PSD) previo al filtrado evidencia características fundamentales del EEG, pero también la presencia de interferencias significativas:

* Se observa un comportamiento típico de decaimiento espectral tipo 1/f, donde las bajas frecuencias concentran la mayor parte de la energía.
* Existe una contribución considerable de la banda delta (0.5–4 Hz), la cual puede dominar el espectro e interferir con la interpretación de bandas funcionales superiores.
* Se identifica un pico pronunciado alrededor de 60 Hz, correspondiente al ruido de la red eléctrica, presente en ambas condiciones.

Adicionalmente, aunque las bandas mu y beta están presentes, su contribución relativa es menos evidente debido al enmascaramiento por componentes de baja frecuencia y ruido.

Esto indica que, en el dominio temporal, las diferencias entre condiciones no son triviales y se encuentran enmascaradas por la complejidad inherente del EEG crudo.

***Etapa Post-filtrado***

Para la etapa de post filtrado se tienen diferentes analisis dependiendo de la Run a analizar:

**Movimiento real (run 3)**

En el dominio temporal se observa una reducción significativa de las oscilaciones de baja frecuencia, evidenciando la eliminación de la deriva de línea base. La señal filtrada presenta una mayor estabilidad y centrado alrededor de cero, manteniendo la morfología de las oscilaciones corticales relevantes.

En el dominio frecuencial, se evidencia:

* Eliminación de componentes por debajo de 0.5 Hz.
* Preservación de las bandas mu y beta.
* Atenuación pronunciada de componentes por encima de 40 Hz.
* Supresión del pico asociado al ruido de red eléctrica.

**Imaginación motora (run 4)**

Se observa un comportamiento consistente con el caso de movimiento real. La señal filtrada presenta:

* Reducción de ruido de baja frecuencia.
* Conservación de la estructura oscilatoria en canales sensorimotores.
* Mejora en la relación señal-ruido.

En el dominio espectral, la PSD filtrada mantiene la distribución de energía en las bandas funcionales, mientras elimina componentes no relevantes, facilitando el análisis posterior de ERD/ERS.

El proceso de filtrado implementado permitió mejorar significativamente la calidad de las señales EEG, eliminando componentes no deseados sin afectar la información relevante para el análisis de actividad motora.

La combinación de un filtro Butterworth de orden moderado, implementación en forma SOS y aplicación mediante filtfilt garantiza:

* Estabilidad numérica
* Ausencia de desfase
* Preservación de amplitudes
* Alta atenuación en bandas no deseadas

Esto establece una base sólida para las etapas posteriores de análisis espectral y comparación entre condiciones de movimiento real e imaginación motora.

Adicionalmente, el análisis comparativo entre las condiciones de movimiento real (run 3) e imaginación motora (run 4) evidencia diferencias consistentes tanto en el dominio temporal como frecuencial, las cuales pueden ser interpretadas a partir de los mecanismos neurofisiológicos asociados al control motor.

**Diferencias en el dominio temporal**

En las señales EEG filtradas, el movimiento real presenta una actividad ligeramente más estructurada y con variaciones de amplitud más marcadas en comparación con la imaginación motora. Esto sugiere una mayor sincronización neuronal asociada a la ejecución efectiva del movimiento, donde intervienen no solo áreas motoras primarias, sino también retroalimentación sensorial (propiocepción).

Por el contrario, en la imaginación motora se observa una señal más difusa y con menor variabilidad de amplitud. Este comportamiento es coherente con la activación parcial de las redes sensorimotoras, donde no existe ejecución física ni retroalimentación periférica, lo que reduce la magnitud de la respuesta cortical.

**Diferencias en el dominio frecuencial**

En la densidad espectral de potencia (PSD), ambas condiciones presentan una distribución global similar (característica tipo 1/f), lo cual es esperable en señales EEG. Sin embargo, se identifican diferencias relevantes:

* En el movimiento real, la energía en las bandas mu (8–12 Hz) y beta (13–30 Hz) tiende a mostrar variaciones más pronunciadas, lo cual se asocia a fenómenos de desincronización relacionada con eventos (ERD) más intensos.

* En la imaginación motora, estas variaciones están presentes pero son de menor magnitud, evidenciando una activación más débil y menos localizada espacialmente.

Este comportamiento coincide con la literatura, donde se reporta que la ERD en movimiento real es más intensa debido a la activación completa del circuito sensorimotor, mientras que en imaginación motora la respuesta es más tenue y dependiente del nivel de entrenamiento del sujeto.
